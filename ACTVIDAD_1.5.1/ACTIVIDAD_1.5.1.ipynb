### Responde en esta celda y a través de código en las siguientes
- Ahora que ya conoces a profundida el dataset realiza IMPUTACIÓN de datos tal como en la tarea 1.5

```python
import numpy as np
df_imput = df.copy()
np.random.seed(42)
for col in df_imput.columns:
    idx = np.random.choice(df_imput.index, 5, replace=False)
    df_imput.loc[idx, col] = np.nan

print("Valores nulos antes de imputar:\n", df_imput.isnull().sum())

df_imput = df_imput.fillna(df_imput.median())

print("\nValores nulos después de imputar:\n", df_imput.isnull().sum())
```

### Responde en esta celda y a través de código en las siguientes
- Elimina outliers si consideras que afectan al estudio

```python
df_clean = df_imput.copy()
Q1 = df_clean.quantile(0.25)
Q3 = df_clean.quantile(0.75)
IQR = Q3 - Q1

is_outlier = ((df_clean < (Q1 - 1.5 * IQR)) | (df_clean > (Q3 + 1.5 * IQR)))
outlier_mask = is_outlier.any(axis=1)

print(f"Filas con outliers detectadas: {outlier_mask.sum()}")

df_clean = df_clean[~outlier_mask].reset_index(drop=True)

print(f"Filas después de eliminar outliers: {df_clean.shape[0]}")
```
